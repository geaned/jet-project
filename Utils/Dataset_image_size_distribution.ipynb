```python
import numpy as np
import matplotlib.pyplot as plt
import os

from collections import Counter
from get_image_size import get_image_size
```

```python
image_widths = []
image_heights = []

IMAGES_PATH = os.path.join(os.path.dirname(os.path.realpath("__file__")), os.pardir, 'datasets_not_split/svhn_all/images')

for file_name in os.listdir(IMAGES_PATH):
    width, height = get_image_size(os.path.join(IMAGES_PATH, file_name))
    image_widths.append(width)
    image_heights.append(height)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(image_widths)
ax2.hist(image_heights)
ax1.title.set_text('Width distribution')
ax2.title.set_text('Height distribution')
plt.show()
```
