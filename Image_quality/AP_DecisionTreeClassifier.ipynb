```python
import pandas as pd

result = pd.read_csv('full_dataset.csv')
result.head()
```

```python
result = result.drop(columns=['Unnamed: 0'])
result.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = result.to_numpy()[:, 2:], list(result.to_numpy()[:, 1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Решающее дерево

Обучим бинарный классификатор на основе решающего дерева:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

params_dict = {
    'max_depth': [3, 4, 5, 6, 7, 8, 9],
    'min_samples_split': [3, 4, 5],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy'],
}

tree = DecisionTreeClassifier()
clf = GridSearchCV(tree, params_dict, cv=5, verbose=10)
clf.fit(X_train, y_train)
print(sorted(clf.cv_results_.keys()))
print(clf.best_params_)

print(clf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier

params_dict = {
    'n_estimators': [2, 3],
    'min_samples_split': [3, 4],
    'max_features': ['sqrt'],
    'criterion': ['gini', 'entropy'],
}

forest = RandomForestClassifier()
clf = GridSearchCV(forest, params_dict, cv=5, verbose=10)
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
print(clf.score(X_test, y_test))
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

# SVM

```python
data = pd.read_csv('full_dataset_normed.csv')
data.head()
```

```python
X, y = data.to_numpy()[:, 2:], list(data.to_numpy()[:, 1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC

params_dict = {
    'C': [1.0, 0.9, 0.8, 1.1, 1.2],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
}

svm_model = SVC()
clf = GridSearchCV(svm_model, params_dict, cv=5, verbose=10)
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
print(clf.score(X_test, y_test))
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
