```python
EXAMPLE_IMG = '/content/drive/MyDrive/recognition_dataset/test_data/IMG_1748_ROD_901074102_0.png'
```

```python
from mmocr.utils.ocr import MMOCR
mmocr = MMOCR(det=None, recog='CRNN')
```

```python
mmocr.readtext(EXAMPLE_IMG, print_result=True, output='./outputs/demo_text_recog_pred.jpg')
```

```python
import os

def predict_img_path(img_path):
  for filename in os.listdir(img_path):
    mmocr.readtext(
        os.path.join(img_path, filename),
        print_result=True,
        output=os.path.join('./outputs', filename),
    )
```

```python
TRAIN_IMAGES = '/content/drive/MyDrive/recognition_dataset/train_data'
TEST_IMAGES = '/content/drive/MyDrive/recognition_dataset/test_data'
```

```python
predict_img_path(TRAIN_IMAGES)
predict_img_path(TEST_IMAGES)
```
