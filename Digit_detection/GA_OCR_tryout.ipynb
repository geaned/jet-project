```python
IMAGE_FOLDER = os.path.join(os.path.dirname(os.path.realpath("__file__")), os.pardir, 'datasets_not_split/rotated_all/images')

reader = easyocr.Reader(['en'])
all_rotated_images = sorted([os.path.join(IMAGE_FOLDER, image_file) for image_file in os.listdir(IMAGE_FOLDER)])
```

```python
for image_path in all_rotated_images:
    current_result = reader.readtext(image_path, allowlist='0123456789-')

    print(image_path.split('/')[-1])
    for guess in current_result:
        best_result_and_confidence = guess[1], guess[2]
        print(best_result_and_confidence)
    print()
```
